Understand the difference between a CDF and a PDF. You're welcome to use code for these calculations. Please include code files in your submission even if you have the code in your report, whether it's .py, .m, .rmd, .ipynb, or whatever. A quick example of PDFs and CDFs in Scipy:

```python
import numpy as np
from scipy.stats import gamma
import matplotlib.pyplot as plt

dist = gamma(a=2, scale=2)
xx = np.linspace(0, 25, 10000)

densities = dist.pdf(xx)

plt.plot(xx, densities)
plt.show()
```

```python
high = 5
prob = dist.cdf(high)

cdf = dist.cdf(xx)
plt.plot(xx, cdf)
plt.axhline(y=prob, color='red')
plt.show()
print(prob)
```

```python
plt.plot(xx, densities)
plt.fill_between(xx, densities, color="blue")
plt.fill_between(xx, densities, where=(xx < high), color="orange")
plt.show()
```

About point masses:

The professor says that your overall prior needs to contain the point mass representing your precise null hypothesis. 

What he means is that you can only test a hypothesis where your prior allows for a probability greater than 0. So if your prior is just a continuous distribution, the probability at any given point is 0. That's why he mixes that point mass and the "spread" distribution, you need a discrete distribution to test a specific point.

At a high level, if your prior doesn't contain your hypothesis, you are essentially ruling out the hypothesis before you even build the rest of your model. You've already predetermined that your hypothesis is impossible, so why are you creating this model?
